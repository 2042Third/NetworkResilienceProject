```python

import numpy as np
import pandas as pd
import csv
import matplotlib
import matplotlib.pyplot as plt  

```

```python

# c_names=["gcc_size","pn"]
# ad = pd.read_csv('dout/out0.010000N1000.csv', names=c_names, header=None)

# file1 = open('dout/out0.010000N1000000.csv', 'r')
# Lines = file1.readlines()
file1 = open('dout/out0.010000N128000.csv', 'r')
Lines = file1.readlines()
count = 0

gcc = []
pn = []
pc = []
err = []
for line in Lines:
    count += 1
    if count == 1:
        continue
    data = [x.strip() for x in line.split(',')]
    gcc.append(int(data[0]))
    pn.append(float(data[1]))
    err.append(0)
    pc.append(float(data[2]))
```

```python
fig1 = plt.figure(1)
plt.errorbar(x=gcc,xerr=err, y=pn,yerr=err, fmt='or', label="Experiement Data")


plt.legend( )#, loc='upper left')
plt.title("pn/p vss gcc")
plt.xlabel('''gcc''')
plt.ylabel("pn/p")

plt.show()
```

```python
fig2 = plt.figure(2)
plt.errorbar(x=gcc,xerr=err, y=pc,yerr=err, fmt='or', label="Experiement Data")


plt.legend( )#, loc='upper left')
plt.title("pc vss gcc")
plt.xlabel('''gcc''')
plt.ylabel("pc")

plt.show()
```
